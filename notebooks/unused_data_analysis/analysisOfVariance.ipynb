```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns


FILENAME = "drug_age.csv"
filepath = f"../data/{FILENAME}"

drug_age_df = pd.read_csv(filepath)

# Display the first few rows of the data
drug_age_df.head()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Read the data into a pandas DataFrame
data = drug_age_df

# Select the relevant columns for the heatmap
columns = ['all drugs', 'all opioids', 'stimulants', 'cannabis', 'benzodiazepine']

# Filter the data for the age categories of interest
age_categories = ['0-15 years']
filtered_data = data[data['age'].isin(age_categories)]

# Calculate the correlation matrix
correlation_matrix = filtered_data[columns].corr()

# Create the heatmap
plt.figure(figsize=(10, 8))  # Adjust the figure size as needed
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.xlabel('Columns')
plt.ylabel('Age Categories')
plt.show()
```

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

# Read the data into a pandas DataFrame
data = drug_age_df

# Select the relevant columns for the heatmap
columns = ['all drugs', 'all opioids', 'stimulants', 'cannabis', 'benzodiazepine']

# Calculate the correlation matrix
correlation_matrix = data.groupby('age')[columns].corr()

# Reshape the correlation matrix
correlation_matrix = correlation_matrix.unstack().reset_index()

# Rename columns
correlation_matrix.columns = ['Column 1', 'Column 2', 'Correlation']

# Pivot the correlation matrix
pivoted_correlation_matrix = correlation_matrix.pivot('Column 1', 'Column 2', 'Correlation')

# Create the heatmap
plt.figure(figsize=(10, 8))  # Adjust the figure size as needed
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.xlabel('Columns')
plt.ylabel('Age Categories')
plt.show()
```

```python
import pandas as pd
import statsmodels.api as sm
from statsmodels.formula.api import ols
data = drug_age_df

# Assuming your data is stored in a DataFrame named 'data'
# Where 'age' is the column for age groups and 'all drugs' is the drug category

# Map '16-34 years' to 1 and '35-54 years' to 0 for the 'age' column
data['age'] = data['age'].map({'16-34 years': 1, '35-54 years': 0})

# Create a formula for the ANOVA model
formula = 'age ~ C(stimulants)'

# Fit the ANOVA model
model = ols(formula, data=data).fit()

# Perform the ANOVA table calculations
anova_table = sm.stats.anova_lm(model, typ=2)

# Print the ANOVA table
print(anova_table)
```

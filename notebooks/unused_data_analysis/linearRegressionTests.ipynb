```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns


FILENAME = "drug_age.csv"
filepath = f"../data/{FILENAME}"

drug_age_df = pd.read_csv(filepath)

# Display the first few rows of the data
drug_age_df.head()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming your data is in csv format and named as 'data.csv'
data = drug_age_df

plt.figure(figsize=(10, 7))

# Create a line plot for each age group
age_plot = sns.lineplot(data=data, x='time', y='all opioids', hue='age')

# Get handles and labels from Seaborn plot
handles, labels = age_plot.get_legend_handles_labels()

# Create the new labels
new_labels = ['0-15 years', '16-34 years', '35-54 years']

# Add legend manually to the right of the plot
plt.legend(handles=handles[1:], labels=new_labels, title='Age Group', bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)

plt.title('All Opioids Usage over Time per Age Group')
# Complete the labels for your plot
plt.xlabel('Time')
plt.ylabel('All Opioids')
plt.grid(True) # This line will add grid lines to the plot.

# Finally, call plt.show() to display the plot.
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming your data is in a DataFrame named data
categories = ['all drugs', 'all opioids', 'stimulants', 'cannabis', 'benzodiazepine']

for category in categories:

    plt.figure(figsize=(10, 7))
  
    # Create a line plot for each age group
    age_plot = sns.lineplot(data=data, x='time', y=category, hue='age')
  
    # Get handles and labels from Seaborn plot
    handles, labels = age_plot.get_legend_handles_labels()
  
    # Create the new labels
    new_labels = ['Age Group', '0-15 years', '16-34 years', '35-54 years']
  
    # Add legend manually to the right of the plot
    plt.legend(handles=handles, labels=new_labels, title=new_labels[0], bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)

    plt.title(f'{category.capitalize()} Usage over Time per Age Group')
    plt.xlabel('Time')
    plt.ylabel(category.capitalize())
    plt.grid(True) # Adds grid lines to the plot for better readability

    # Displays the plot
    plt.show()
```
